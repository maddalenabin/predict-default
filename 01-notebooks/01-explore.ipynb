## Define and look at target variable

The target variable is whether a customer defaults on their payment within 21 days. We can create this target variable based on the `amount_outstanding_21d` column, where a value greater than 0 indicates a default.

```python
df["default"] = (df['amount_outstanding_21d'] > 0).astype(int) 
# true 1 if defaulted, false 0 if not
```

```python
print(df["default"].value_counts(normalize=True)*100)
```

```python
# Calculate default rate
default_rate = df['default'].mean()
n_defaults = df['default'].sum()
n_total = len(df)

print(f"Default rate: {default_rate:.3f} ({default_rate*100:.1f}%)")
print(f"Defaults: {n_defaults:,} out of {n_total:,} loans.")
```

```python
target = "default"  

# class imbalance
sns.countplot(x=target, data=df, palette=["#e16d07","#0d06d1"])
plt.title("Target distribution (default vs non-default)")
plt.show()

# default rate over time
df["loan_issue_date"] = pd.to_datetime(df["loan_issue_date"])
df.groupby(df["loan_issue_date"].dt.to_period("M"))[target].mean().plot(marker="o")
plt.title("Monthly default rate over time")
plt.ylabel("Default rate")
plt.show()
```

## Missing values

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

Drop loan entries where "card_expiry_month" and "card_expiry_year" is missing.

```python
df = df.dropna(subset=["card_expiry_month", "card_expiry_year"])
```

Fill missing number with 0. Here I am assuming that the missing values correspond to the fact that "existing_klarna_debt" is zero whether is not present. It is anyways unknown, so we cannot assume there is a dept in place.

```python
df["existing_klarna_debt"].fillna(0, inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
df[numeric_cols] = df[numeric_cols].fillna(0)
```

# Feature distributions

```python
#numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# histogram for a few numeric features
df[numeric_cols].hist(figsize=(15,10), bins=30)
plt.tight_layout()
plt.show()
```

# Correlation

```python
# Correlations with target
corr_matrix = df[numeric_cols + [target]].corr()
corrs = corr_matrix[target].drop(target)  # drop target self-correlation = 1.0

corr_matrix
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0, annot=False)
plt.title("Correlation matrix heatmap")
plt.show()
```

```python
corrs = df[numeric_cols].corrwith(df["default"])
#print(corrs.sort_values(ascending=False))

corrs.drop("default").sort_values().plot(kind="bar", figsize=(5,4))
plt.title(f"Correlation of features with {target}")
plt.show()
```

**Observatiuons:**
- `days_since_first_loan`: negatively correlated with default, meaning that customers with a longer history with Klarna are less likely to default. Though, the correlation absolute value is small. 
- `num_failed_payments_3m`, `num_failed_payments_6m`, `num_failed_payments_1y`: positively correlated with default, meaning that customers who have failed payments in the last 3/6 months or 1 yr are more likely to default.
- `card_expiry_year` is also positively correlated, which I didn't expect and could just be spourious correaltion.
- We have only loans given during Jun-Oct. So no sample data on the other months. This can be a limitation when predicting loans to be issued on the missing months.
